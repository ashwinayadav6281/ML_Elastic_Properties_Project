```python
from matminer.featurizers.conversions import StrToComposition
df = StrToComposition().featurize_dataframe(df, "formula")
df.head()
```

```python
df["bulk_modulus_vrh"] = df["bulk_modulus"].apply(lambda x: x.get("vrh") if isinstance(x, dict) else None)
```

```python
y = df['bulk_modulus_vrh'].values

# Get the current columns in the DataFrame
current_columns = df.columns.tolist()

# Define the columns you want to exclude from the feature set X
# These are the columns you want to drop from the original DataFrame
columns_to_exclude = ["material_id", "formula", "structure", "bulk_modulus"]

# Create a list of columns to drop from the current DataFrame
# Ensure that only columns that exist in the current_columns are included in the drop list
unwanted_columns_for_X = [col for col in current_columns if col in columns_to_exclude]

# Drop the unwanted columns to get the feature set X
X = df.drop(unwanted_columns_for_X, axis=1)

print("There are {} possible descriptors:\n\n{}".format(X.shape[1], X.columns.values))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
from matminer.featurizers.composition import ElementProperty
from sklearn.model_selection import KFold, cross_val_score
import pandas as pd

# Drop rows with NaN values in either X or y
# We concatenate X and y to drop rows where _any_ value is NaN
df_for_fitting = pd.concat([X, pd.Series(y, name='bulk_modulus_vrh')], axis=1).dropna()

# Separate X and y again from the cleaned DataFrame
X_cleaned = df_for_fitting.drop('bulk_modulus_vrh', axis=1)
y_cleaned = df_for_fitting['bulk_modulus_vrh'].values

print(f"Original dataset size: {len(df)}")
print(f"Dataset size after dropping rows with NaNs: {len(df_for_fitting)}")

print("There are {} possible descriptors in cleaned data:\n\n{}".format(X_cleaned.shape[1], X_cleaned.columns.values))


lr = LinearRegression()

# Use the cleaned data for fitting and cross-validation
lr.fit(X_cleaned, y_cleaned)

# get fit statistics
print('Linear Regression results:')
print('training R2 = ' + str(round(lr.score(X_cleaned, y_cleaned), 3)))
print('training RMSE = %.3f' % np.sqrt(mean_squared_error(y_true=y_cleaned, y_pred=lr.predict(X_cleaned))))

# Use 10-fold cross validation (90% training, 10% test)
crossvalidation = KFold(n_splits=10, shuffle=True, random_state=1)
# Use the cleaned data for cross-validation
scores = cross_val_score(lr, X_cleaned, y_cleaned, scoring='neg_mean_squared_error', cv=crossvalidation, n_jobs=1)
rmse_scores = [np.sqrt(abs(s)) for s in scores]
r2_scores = cross_val_score(lr, X_cleaned, y_cleaned, scoring='r2', cv=crossvalidation, n_jobs=1)

print('Cross-validation results:')
print('Folds: %i, mean R2: %.3f' % (len(scores), np.mean(np.abs(r2_scores))))
print('Folds: %i, mean RMSE: %.3f' % (len(scores), np.mean(np.abs(rmse_scores))))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=50, random_state=1)

df_for_fitting = pd.concat([X, pd.Series(y, name='bulk_modulus_vrh')], axis=1).dropna()

# Separate X and y again from the cleaned DataFrame
X_cleaned = df_for_fitting.drop('bulk_modulus_vrh', axis=1)
y_cleaned = df_for_fitting['bulk_modulus_vrh'].values

rf.fit(X_cleaned, y_cleaned)
print('training R2 = ' + str(round(rf.score(X_cleaned, y_cleaned), 3)))
print('training RMSE = %.3f' % np.sqrt(mean_squared_error(y_true=y_cleaned, y_pred=rf.predict(X_cleaned))))
```
